シンプルなoneHotで試してみる．

```python
import numpy as np
import pandas as pd
```

```python
d = 10000
n = 1000
```

```python
cohort_roots = [np.random.binomial(100, 0.001, d) for i in range(20)]
result = []
for i in range(0, n):
  # 誤差乗せる
  root_idx = i % len(cohort_roots)
  result.append(cohort_roots[root_idx] + np.random.binomial(1, 0.001, d)) 
  #誤差 ０.01だとそこそこブレる， まぁぶれたならそこが大事だからちゃんと反映されるねという話はある
```

```python
weights = [np.random.randint(-1, 2, d) for i in range(100)] # 適当に50bitでやってみる
```

```python
vecs = []
for i in range(len(result)):
  vec = []
  for weight in weights:
    vec.append(1 if np.dot(result[i], weight) >= 0 else 0)
  vecs.append(vec)
```

```python
bites = [bidnizer(each) for each in vecs]
```

```python
def bidnizer(tgt : list):
  out = 0
  for each in tgt:
    out = (out << 1) | each
  return out
```

```python
cohort_id = []
for i in range(0, n):
  root_idx = i % len(cohort_roots)
  cohort_id.append(root_idx)
```

```python
tmpDf = pd.DataFrame(list(zip(bites, cohort_id)), columns = ["bite", "cohort_id"])
```

```python
sTmpDf = tmpDf.sort_values("bite")
```

```python
for each in sTmpDf.cohort_id:
  print(each)
# そこそこ分類出来ていることがわかる， あっているかは知らん
```
